### 5. Lấy danh sách kho (Stocks)

```python
import pandas as pd
import requests

# URL API để lấy danh sách kho
url_stocks = "https://crmconnect.misa.vn/api/v2/Stocks"

# Gửi yêu cầu GET
response_stocks = requests.get(url_stocks, headers=headers)

if response_stocks.status_code == 200:
    stocks_data = response_stocks.json()
    
    # Chuyển đổi dữ liệu JSON thành DataFrame
    df_stocks = pd.DataFrame(stocks_data['data'])
    
    # Tùy chỉnh hiển thị để xem toàn bộ cột và dòng
    pd.set_option('display.max_columns', None)  # Hiển thị tất cả các cột
    pd.set_option('display.max_rows', None)  # Hiển thị tất cả các dòng nếu cần
    display(df_stocks)  # Nếu dùng Jupyter Notebook
else:
    print(f"Lỗi khi lấy dữ liệu kho: {response_stocks.status_code} - {response_stocks.text}")
```
